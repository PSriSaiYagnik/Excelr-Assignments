```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_csv('Patient_Data.csv')
df2 = pd.read_csv('Billing_Data.csv')
```

```python
### Task 2
billing_relevant = df1[['PatientID', 'Department', 'Doctor', 'BillAmount']]
print("\nBilling Relevant Columns:\n")
billing_relevant.head()
```

```python
### Task 3
patients_cleaned = df1.drop(columns=['ReceptionistID', 'CheckInTime'], errors='ignore')
print("\nAfter Dropping Admin Columns:\n")
patients_cleaned.head()
```

```python
### Task 4
dept_revenue = billing_relevant.groupby('Department')['BillAmount'].sum()
print("\nTotal Bill per Department:\n")
dept_revenue
```

```python
### Task 5
patients_unique = patients_cleaned.drop_duplicates(subset=['PatientID'])
print("\nUnique Patients:\n")
patients_unique
```

```python
### Task 6
mean_bill = billing_relevant['BillAmount'].mean()
billing_relevant['BillAmount'].fillna(mean_bill, inplace=True)
print("\nMissing BillAmount Filled with Mean:\n")
billing_relevant
```

```python
### Task 7
merged_data = pd.merge(patients_unique, df2, on='PatientID', how='inner')
print("\nMerged Dataset:\n")
merged_data.head()
```

```python
#Task 8
new_patients = pd.DataFrame({
    "Name" : ["Sri", "Sai"],
    "PatientID": [201, 202],
    "Department": ["Cardiology", "Neurology"],
    "Doctor": ["Dr. Steve", "Dr. Mary"],
    "BillAmount": [2000, 3500]
})
patients_extended = pd.concat([patients_unique, new_patients], ignore_index=True)
print("\nExtended Patients Dataset:\n")
patients_extended.head(7)
```

```python
# Task 9
new_columns = pd.DataFrame({
    "InsuranceCovered": (["Yes", "No", "Yes", "No"] * (len(patients_extended) // 4 + 1))[:len(patients_extended)],
    "FinalAmount": patients_extended['BillAmount'] * 0.9
})
final_dataset = pd.concat([patients_extended.reset_index(drop=True), new_columns], axis=1)
print("\nFinal Cleaned Dataset:\n")
final_dataset.head()
```
